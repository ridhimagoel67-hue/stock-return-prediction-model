```python
t1 = pd.read_csv("../data/raw/t1_prices.csv")

t1.head()
```

```python
balance[['join_key', 'Current Price']].head()
```

```python
master = (
    balance
    .merge(ratios, on="join_key", suffixes=("_bal", "_rat"))
    .merge(t1, on="join_key")
)

master.shape
```

```python
master.head()
```

```python
master.info()
```

```python
master["future_return"] = (
    (master["t_1_price"] - master["Current Price_bal"])
    / master["Current Price_bal"]
)
```

```python
master["future_return"].describe()
```

```python
def label_return(x):
    if x > 0.05:
        return "Buy"
    elif x < -0.05:
        return "Sell"
    else:
        return "Hold"

master["signal"] = master["future_return"].apply(label_return)

master["signal"].value_counts()
```

```python
df = master.copy()

drop_cols = [
    "Name_bal",
    "BSE Code_bal",
    "NSE Code_bal",
    "join_key"
]

df = df.drop(columns=drop_cols)

df.shape
```
