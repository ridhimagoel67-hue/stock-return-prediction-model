```python
import pandas as pd
import numpy as np

df = pd.read_csv("../../data/processed/master.csv")

print(df.shape)
df.head()
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns

print(len(numeric_cols))
```

```python
df.select_dtypes(exclude=np.number).columns.tolist()
```

```python
target_rows = 50000

sampled_df = df.sample(
    n=target_rows,
    replace=True,
    random_state=42
).reset_index(drop=True)

print(sampled_df.shape)
```

```python
np.random.seed(42)

for col in numeric_cols:

    std = sampled_df[col].std()

    noise = np.random.normal(
        loc=0,
        scale=std * 0.02,
        size=len(sampled_df)
    )

    sampled_df[col] = sampled_df[col] + noise
```

```python
print(sampled_df.shape)

sampled_df.head()
```

```python
sampled_df.to_csv(
    "../data/big_stock_dataset_50k.csv",
    index=False
)

print("Dataset saved successfully")
```

```python
sampled_df.shape
sampled_df.memory_usage(deep=True).sum() / 1024**2
```

```python
target_rows = 500000

big_df = df.sample(
    n=target_rows,
    replace=True,
    random_state=42
).reset_index(drop=True)

print(big_df.shape)
```

```python
np.random.seed(42)

for col in numeric_cols:

    std = big_df[col].std()

    noise = np.random.normal(
        loc=0,
        scale=std * 0.02,
        size=len(big_df)
    )

    big_df[col] = big_df[col] + noise
```

```python
print(big_df.shape)

memory_mb = (
    big_df.memory_usage(deep=True).sum()
    / 1024**2
)

print(f"Memory Usage: {memory_mb:.2f} MB")
```

```python
big_df.to_csv(
    "../data/big_stock_dataset_500k.csv",
    index=False
)

print("500K dataset saved")
```
